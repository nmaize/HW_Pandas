```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])

#Display first 5 rows of data
school_data_complete.head()
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the overall passing rate (overall average score), i.e. (avg. math score + avg. reading score)/2

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
total_school = school_data_complete["school_name"].nunique()
total_students = school_data_complete["Student ID"].count()      #Total sum # is not the same
total_budget = school_data_complete["budget"].unique().sum()    #Method Chaining #Add $ and 2 decimal places
avg_math_score = school_data_complete["math_score"].mean()
avg_reading_score = school_data_complete["reading_score"].mean()
perct_math_passing=school_data_complete.loc[school_data_complete["math_score"] >=70].count()/total_students*100
perct_read_passing=school_data_complete.loc[school_data_complete["reading_score"] >=70].count()/total_students*100
overall_passing=(avg_math_score+avg_reading_score)/2
school_type=school_data_complete["type"].unique()
per_student_budget = total_budget/total_students
                                              
```

```python
#Define DataFrame

#school_data_complete.dtypes
district_summary_df = pd.DataFrame({
    "Total Schools": [total_school],
    "Total Students": total_students,
    "Total Budget": [total_budget],
    "Average Math Score": [avg_math_score],
    "Average Reading Score": [avg_reading_score],
    "% Passing Math": [perct_math_passing[2]],
    "% Passing Reading": [perct_read_passing[2]],
    "% Overall Passing Rate":[overall_passing]
})
district_summary_df

#Calculate/Count the total number of schools
#Total Schools['school_name'].value_counts()
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)
  
* Create a dataframe to hold the above results

```python
#Define DataFrame

# Create group and add index as school name
name_of_school = school_data_complete.groupby(['school_name'])

    
name_of_school.count()
```

```python
#School Type
school_type = name_of_school['type'].unique() 

#Total Students
school_student_count = name_of_school['student_name'].count()

#budget
school_budget = name_of_school['budget'].mean()

#per student budget
per_student_budget = school_budget/school_student_count

#average math score
gavg_math_score = name_of_school['math_score'].mean()

#average reading score
gavg_reading_score = name_of_school['reading_score'].mean()

#percent math - Created math_group
#percent_math = name_of_school['math_score'].values() #>=70].count()/school_student_count*100
mpassing_score_df = school_data_complete[(school_data_complete["math_score"]>70)]
math_group = mpassing_score_df.groupby(['school_name']).count()
math_group = math_group['math_score']/school_student_count*100


#percent passing reading - Created reading_group
#percent_reading = name_of_school['reading_score'].values()
rpassing_score_df = school_data_complete[(school_data_complete["reading_score"]>70)]
reading_group = rpassing_score_df.groupby(['school_name']).count()
reading_group = reading_group['reading_score']/school_student_count*100

#Overall Passing Rate
overall_passing = (math_group + reading_group) / 2
```

```python
#Create a grouped by DF on school and school type
summary_school_df = pd.DataFrame(data={
    
   "School Type": school_type,
   "Total Students": school_student_count,
   "Total School Budget": school_budget,
   "Per Student Budget": per_student_budget, 
   "Average Math Score": gavg_math_score,
   "Average Reading Score": gavg_reading_score,
   "% Passing Math": math_group,
   "% Passing Reading": reading_group,
   "% Overall Passing Rate":overall_passing
})


#Convert list from list and re-assign in summary school DF
convert= np.hstack(summary_school_df['School Type'])

summary_school_df['School Type'] = convert
summary_school_df.head()
```
